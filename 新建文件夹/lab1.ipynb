```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats

x_min = -4.0
x_max = 4.0
mean = 0.0
std = 1.0

x = np.linspace(x_min, x_max, 100000)

#x_dist = stats.norm.rvs(loc = 0, scale = 1, size = 10000)
#x_cdf = scipy.stats.norm.cdf(loc = 0, scale = 1, size = 1000)
#dgraph = stats.norm.pdf(0.1, 0, 1)

a = -2
b = -0.5
c = 0
d = 2.5
```

```python
a_sigma = stats.norm.cdf(a)
b_sigma = stats.norm.cdf(b)
c_sigma = stats.norm.cdf(c)
d_sigma = stats.norm.cdf(d)

a2 = stats.norm.ppf(a_sigma)
b2 = stats.norm.ppf(b_sigma)
c2 = stats.norm.ppf(c_sigma)
d2 = stats.norm.ppf(d_sigma)

print(a)
print(b)
print(c)
print(d)
print()
print(a_sigma)
print(b_sigma)
print(c_sigma)
print(d_sigma)
print()
print(a2)
print(b2)
print(c2)
print(d2)
```

Now we have input values of -2, -0.5, 0, and 2.5. After integrated in the distribution and then using sigma to get the coresponding values again, we have -2.000000000000001, -0.5000000000000001, 0.0, 2.4999999999999987, which are with little differencec to the original values.
